```python
#! /usr/bin/env python
# Author : Manas Mahale <manas.mahale@bcp.edu.in>


import pandas as pd
import numpy as np
import pylab as plt
from rdkit.Chem.Descriptors import *
from rdkit import Chem
from mordred import Calculator, descriptors
from multiprocessing import freeze_support

%matplotlib inline
```

```python
data = pd.read_csv("../Data/external/Dataset_1.tsv", sep='\t')
data = data.filter(['Ligand SMILES', 'IC50 (nM)'])
```

```python
lig, ic50 = data['Ligand SMILES'].to_list(), data['IC50 (nM)'].to_numpy()
```

```python
if __name__ == "__main__":
    freeze_support()
    mols = [Chem.MolFromSmiles(i) for i in lig]
    calc = Calculator(descriptors)
    a = calc.pandas(mols)
```

```python
a
```

```python
columns = a.columns
```

```python
for i in columns:
    if a[i].dtype == 'O':
        a = a.drop(i, axis=1)
```

```python
a.var().describe()
```

```python
variance = a.var()
```

```python
variable = []

for i in range(0,len(variance)):
    if variance[i]>=1.091364e-04: #setting the threshold as 1%
        variable.append(a.columns[i])
```

```python
variable
```

```python
a[variable]
```

```python
a[variable].to_csv('../Data/processed/descriptors.csv')
```
